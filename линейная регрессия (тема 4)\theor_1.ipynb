```python
import numpy as np 
from  scipy import stats
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.sandbox.regression.predstd import wls_prediction_std

sigma = [0.1,1,3]
N = [10,100,1000]
xx = [[0.5],[2],[25]]
a=1
for n in N:
  for s in sigma:
    X = stats.uniform.rvs(0,1,size=n)
    Y = a*X + stats.norm.rvs(0,np.sqrt(s),size=n)
    model = sm.OLS(Y, X)
    #print(predict.summary())
    fig, ax = plt.subplots(figsize=(8,6))
    ax.plot(X, Y, 'o', label="data")
    ax.plot(X, model.fit().fittedvalues, label="OLS",c='green')
    ax.plot(sorted(X),a*sorted(X), c = 'yellow')
    pr=[]
    for x in xx:
      pr.append('y_predict='+str(round(model.predict(params=model.fit().params, exog=x), 2)) +'     y='+str(a*x) )
    plt.title('n='+ str(n)+'    sigma='+str(s)+'\n' +pr[0]+'\n' +pr[1]+'\n' +pr[2] )
    print( model.fit().summary() )
    print('\n')
```
